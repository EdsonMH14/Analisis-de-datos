```
#Importamos las librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Cargar el archivo CSV
df= pd.read_csv('global_inflation_countries.csv') #Este data set fue extraido de Kaggle

# Mostrar las primeras filas del DataFrame
df.head()
```

```
# La salida es de la forma (filas, columnas) del data frame.
df.shape
```

```
# Es un método que muestra un resumen conciso del DataFrame
df.info()
```

```
# Método que calcula estadísticas descriptivas para las columnas numéricas del DataFrame
#count: Número de valores no nulos.
#mean: Media (promedio) de los valores.
#std: Desviación estándar (medida de dispersión).
#min: Valor mínimo.
#25%: Primer cuartil (25% de los datos están por debajo de este valor).
#50%: Mediana (50% de los datos están por debajo de este valor).
#75%: Tercer cuartil (75% de los datos están por debajo de este valor).
#max: Valor máximo.
df.describe()
```

```
# Cuenta los valores nulos en cada columna del DataFrame
df.isnull().sum()
```

```
#LIMPIEZA DE LOS DATOS FALTANTES

# Rellena los valores nulos en la columna 'region' con el texto 'Desconocido'
df['region'].fillna('Desconocido', inplace=True)

df['intermediate_region'].fillna('Desconocido', inplace=True)
```

```
# Mostrar las primeras filas del DataFrame
df.head()
```

```
df.tail()
```

```
# Contar los paises unicos
num_paises_unicos = df['country_name'].nunique()
print(f"Número de países únicos: {num_paises_unicos}")
```

```
# Filtrar el DataFrame para el año 2023 por sub-region
df_2023 = df[df['year'] == 2023]

plt.figure(figsize=(12, 8))
plt.bar(df_2023['sub_region'], df_2023['inflation_rate'], color='skyblue')
plt.title('Tasa de Inflación en 2023 por País')
plt.xlabel('País')
plt.ylabel('Tasa de Inflación (%)')
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x para mejor legibilidad
plt.grid(True, axis='y')  # Añadir cuadrícula solo en el eje y
plt.show()
```

```
# Accdemos a los datos de mexico
datos_mexico = df[df['country_name'] == 'MEXICO']
print(datos_mexico)
```

```
# Encontrar el año con mas inflacion en mexico
plt.figure(figsize= (10, 6))
plt.plot(datos_mexico['year'], datos_mexico['inflation_rate'], marker='o', linestyle='-')
plt.title('Tasa de Inflación en México a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Tasa de Inflación (%)')
plt.grid(True)
plt.show()
```

```
# ENCONTRAR EL AÑO CON LA MAYOR INFLACION EN MEXICO

# Encontrar el índice de la fila con la mayor inflación
indice_max_inflacion = df_mexico['inflation_rate'].idxmax()

# Obtener la fila correspondiente al año con la mayor inflación
anio_max_inflacion = df_mexico.loc[indice_max_inflacion, 'year'] #  Accede al valor de la columna 'year' en la fila con la mayor inflación.
inflacion_max = df_mexico.loc[indice_max_inflacion, 'inflation_rate'] # Accede al valor de la columna 'inflation_rate' en la fila con la mayor inflación.

print(f"El año con la mayor inflación en México fue {anio_max_inflacion} con una tasa de inflación del {inflacion_max:.2f}%.")
```

```
# ENCONTRAR EL AÑO CON LA MENOR INFLACION EN MEXICO

# Encontrar el índice de la fila con la mayor inflación
indice_max_inflacion = df_mexico['inflation_rate'].idxmin()

# Obtener la fila correspondiente al año con la mayor inflación
anio_max_inflacion = df_mexico.loc[indice_max_inflacion, 'year'] #  Accede al valor de la columna 'year' en la fila con la mayor inflación.
inflacion_max = df_mexico.loc[indice_max_inflacion, 'inflation_rate'] # Accede al valor de la columna 'inflation_rate' en la fila con la mayor inflación.

print(f"El año con la menor inflación en México fue {anio_max_inflacion} con una tasa de inflación del {inflacion_max:.2f}%.")
```

```
# AGRUPAR POR PAÍS Y CALCULAR EL PROMEDIO DE INFLACIÓN

# Calcular el promedio de inflación histórica por país
promedio_inflacion_por_pais = df.groupby('country_name')['inflation_rate'].mean().reset_index()

# Renombrar la columna del promedio para mayor claridad
promedio_inflacion_por_pais = promedio_inflacion_por_pais.rename(columns={'inflation_rate': 'promedio_inflacion'})

# Mostrar el DataFrame resultante
print(promedio_inflacion_por_pais)
```

```
# ORDENAR Y SELECCIONAR LOS 10 PAÍSES CON MAYOR INFLACIÓN

# Ordenar por promedio de inflación de mayor a menor y seleccionar los 10 primeros
top_10_paises = promedio_inflacion_por_pais.sort_values(by='promedio_inflacion', ascending=False).head(10)

print(top_10_paises)
```

```
# GRAFICAR LOS 10 PAÍSES CON MAYOR INFLACIÓN

plt.figure(figsize=(12, 6))
sns.barplot(x='country_name', y='promedio_inflacion', data=top_10_paises, palette='Reds_r')

# Guardamos el objeto del gráfico en la variable ax. Esto nos permite acceder a las propiedades del gráfico, como las barras (patches).
ax = sns.barplot(x='country_name', y='promedio_inflacion', data=top_10_paises, palette='Reds_r')

# Añadir los valores de cada barra
for p in ax.patches:
    height = p.get_height()  # Altura de la barra (frecuencia)
    ax.text(
        p.get_x() + p.get_width() / 2.,  # Posición X: centro de la barra
        height + 0.2,                     # Posición Y: altura de la barra + un pequeño offset
        f'{int(height)}',                  # Texto: valor de la frecuencia
        ha='center',                       # Alineación horizontal: centrada
        va='bottom'                        # Alineación vertical: en la parte inferior del texto
    )

plt.title('Top 10 Países con Mayor Inflación Promedio (Historico)')
plt.xlabel('País')
plt.ylabel('Promedio de Inflación (%)')
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x para mejor legibilidad
plt.grid(True, axis='y')  # Añadir cuadrícula solo en el eje y
plt.show()
```

```
# ORDENAR Y SELECCIONAR LOS 10 PAÍSES CON MENOR INFLACIÓN

# Ordenar por promedio de inflación de mayor a menor y seleccionar los 10 primeros
top_10_paises = promedio_inflacion_por_pais.sort_values(by='promedio_inflacion', ascending=True).head(10)

print(top_10_paises)
```

```
# ENCONTRAR LA FILA CON LA MAYOR INFLACIÓN

# Encontrar el índice de la fila con la mayor inflación
indice_max_inflacion = df['inflation_rate'].idxmax()

# Obtener la fila correspondiente al máximo valor de inflación
fila_max_inflacion = df.loc[indice_max_inflacion]

# Extraer el país, el año y la inflación máxima
pais_max_inflacion = fila_max_inflacion['country_name']
anio_max_inflacion = fila_max_inflacion['year']
inflacion_max = fila_max_inflacion['inflation_rate']

print(f"El país con la mayor inflación en la historia es {pais_max_inflacion} en el año {anio_max_inflacion} con una tasa de inflación del {inflacion_max:.2f}%.")
```
